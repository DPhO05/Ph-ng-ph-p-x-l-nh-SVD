```python
import numpy as np
import matplotlib.pyplot as plt

def compress_image_per_channel(image):
    # Tách các kênh R, G, B từ ảnh
    R, G, B = image[:, :, 0], image[:, :, 1], image[:, :, 2]
    
    # Khởi tạo các danh sách chứa ma trận SVD cho từng kênh
    channels = [R, G, B]
    channel_names = ['Red', 'Green', 'Blue']
    
    # Thiết lập đồ họa
    fig, axes = plt.subplots(3, 3, figsize=(12, 12))
    
    for idx, channel in enumerate(channels):
        # Áp dụng SVD cho từng kênh
        U, S, V = np.linalg.svd(channel, full_matrices=False)
        
        # Chuẩn hóa các ma trận U, S, V
        U_rescaled = (U - np.min(U)) / (np.max(U) - np.min(U))
        S_matrix = np.diag(S)  # Chuyển S thành ma trận đường chéo
        S_rescaled = (S_matrix - np.min(S_matrix)) / (np.max(S_matrix) - np.min(S_matrix))
        V_rescaled = (V - np.min(V)) / (np.max(V) - np.min(V))
        
        # Hiển thị U, S, V cho kênh hiện tại
        matrices = [U_rescaled, S_rescaled, V_rescaled]
        titles = [f"{channel_names[idx]} - U", f"{channel_names[idx]} - S", f"{channel_names[idx]} - V"]
        
        for i in range(3):
            axes[idx, i].imshow(matrices[i], cmap='gray')
            axes[idx, i].set_title(titles[i])
            axes[idx, i].axis('off')
    
    plt.tight_layout()
    plt.show()

# Giả sử 'image' là ảnh RGB (3 kênh màu)
# compress_image_per_channel(image)
```

```python
def compress_image(image):
    U, S, V = np.linalg.svd(image, full_matrices = False)
     # Chuẩn hóa ma trận U về khoảng [0, 1]
    S_matrix = np.zeros((U.shape[0], V.shape[0]))  # Tạo ma trận 0 với kích thước tương ứng
    np.fill_diagonal(S_matrix, S)  # Điền các giá trị S vào đường chéo
    S_rescaled = (S_matrix - np.min(S_matrix)) / (np.max(S_matrix) - np.min(S_matrix))
    
    U_rescaled = (U - np.min(U)) / (np.max(U) - np.min(U))
    V_rescaled = (V - np.min(V)) / (np.max(V) - np.min(V))
    matrices = [U_rescaled, S_rescaled, V_rescaled]
    titles = ["U", "S", "V"]
    fig = plt.figure(figsize = (10,10))
    for i in range(len(matrices)):
        plt.subplot(1,3,i+1)
        plt.imshow(matrices[i], cmap='gray')
        plt.title(titles[i])
    plt.show()
    
```
